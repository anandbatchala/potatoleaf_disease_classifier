```python
training_size=0.8
validation_size=0.1
test_size=0.1
```

```python
def get_dataset_partitions_tf(ds, train_split=0.8, val_split=0.1, test_split=0.1, shuffle=True, shuffle_size=10000):
  ds_size = len (ds)
  if shuffle:
    ds = ds.shuffle(shuffle_size, seed=12)
  train_size = int(train_split*ds_size)
  val_size = int(val_split * ds_size)
  train_ds = ds.take (train_size)
  val_ds = ds.skip(train_size).take(val_size)
  test_ds = ds.skip(train_size).skip(val_size)
  return train_ds, val_ds, test_ds
```

```python
train_ds, val_ds, test_ds= get_dataset_partitions_tf(dataset)
```

```python
train_ds = train_ds.cache().shuffle (1000). prefetch(buffer_size=tf.data.AUTOTUNE)
val_ds = val_ds.cache().shuffle (1000). prefetch(buffer_size=tf.data.AUTOTUNE)
test_ds = test_ds.cache().shuffle (1000). prefetch (buffer_size=tf.data.AUTOTUNE)
```

```python
from tensorflow.keras import layers
from tensorflow.keras import Sequential

IMAGE_SIZE = 256  # Example size, adjust as needed

resize_and_rescale = Sequential([
    layers.Resizing(IMAGE_SIZE, IMAGE_SIZE),
    layers.Rescaling(1.0 / 255)
])
```

```python
data_augmentation = Sequential([
    layers.RandomFlip("horizontal_and_vertical"),
    layers.RandomRotation(0.2)
])
```

```python
IMAGE_SIZE = 256
BATCH_SIZE = 32
CHANNELS =3
```

```python
from tensorflow.keras import layers, models

# Assuming resize_and_rescale and data_augmentation are pre-defined preprocessing layers
input_shape = (BATCH_SIZE, IMAGE_SIZE, IMAGE_SIZE, CHANNELS)
n_classes = 3

# Adjust the input shape for the Input layer (excluding BATCH_SIZE)
input_layer_shape = (IMAGE_SIZE, IMAGE_SIZE, CHANNELS)

model = models.Sequential([
    layers.Input(shape=input_layer_shape),  # Define input shape here
    resize_and_rescale,
    data_augmentation,
    layers.Conv2D(32, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, kernel_size=(3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, kernel_size=(3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(n_classes, activation='softmax'),
])

# Remove model.build(input_shape=input_shape) as it's unnecessary with Input layer
```

```python
model.summary()
```

```python
model.compile(
optimizer='adam',
loss=tf.keras. losses. SparseCategoricalCrossentropy (from_logits=False),
metrics=['accuracy']
)
```

```python
history = model.fit(
train_ds,
epochs=50,
batch_size=BATCH_SIZE,
verbose=1,
validation_data=val_ds
)
```

```python
score = model.evaluate(test_ds)
```
